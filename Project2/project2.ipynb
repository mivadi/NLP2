## Read data

```python
def read_data(names, en_words = [], fr_words = []):
    
    for name in names:
        if name == names[0]:
            with open(name) as f:
                en_tr = codecs.open(name,'r', encoding='utf-8-sig').readlines()
        if name == names[1]:
                fr_tr = codecs.open(name,'r', encoding='utf-8-sig').readlines()

    # Create dataframe
    data = pd.DataFrame({'en': en_tr,'fr': fr_tr})

    # Get strings and delete \n 
    for i in data.index:
        if '\n' in data.at[i, 'en']:
            data.at[i, 'en'] = data.at[i, 'en'].replace('\n','')
        if '\n' in data.at[i, 'fr']:
            data.at[i, 'fr'] = data.at[i, 'fr'].replace('\n','')
                
    # Get all bilingual sentences
    sentences = []
    
    for x,y in zip(data['en'], data['fr']):
        sentences.append([x,y])
        
    for s in sentences:
        for i in range(len(s[0].split())):
            en_words.append(s[0].split()[i])
        for i in range(len(s[1].split())):
            fr_words.append(s[1].split()[i])

        en_words, fr_words = sorted(list(set(en_words))) , sorted(list(set(fr_words)))
            
    print('Size of English dictionary:', len(en_words))
    print('Size of French dictionary:', len(fr_words))  
           
    return data, sentences, en_words, fr_words
```

```python
class language:
    def __init__(self, name):
        self.name = name
        self.word2index = {}
        self.word2count = {}
        self.index2word = {0: "SOS", 1: "EOS"}
        self.n_words = 2  # Count SOS and EOS

    def addSentence(self, sentence):
        for word in sentence.split(' '):
            self.addWord(word)

    def addWord(self, word):
        if word not in self.word2index: #if new word
            self.word2index[word] = self.n_words #get index of words in dictionary
            self.word2count[word] = 1 #get count of words in dictionary
            self.index2word[self.n_words] = word
            self.n_words += 1
        else:
            self.word2count[word] += 1
```

```python
def sentencesToTensor(pair, input_lang, output_lang):
    
    EOS_token = 1
    
    indexes = [input_lang.word2index[word] for word in pair[1].split(' ')]
    indexes.append(EOS_token)
    input_tensor = torch.tensor(indexes, dtype=torch.long).view(-1, 1)
 
    indexes = [output_lang.word2index[word] for word in pair[0].split(' ')]
    indexes.append(EOS_token)
    target_tensor = torch.tensor(indexes, dtype=torch.long).view(-1, 1)

    return (input_tensor, target_tensor)
```

## Natural Machine Translation (NMT) model

The NMT model exists of a encoder and decoder. Below the encoder and decoder are defined. They are combined in the NMTmodel class.

```python
class Encoder(nn.Module):

    def __init__(self, source_dim, RNN):
        """
        Initialize Encoder.
            :param source_dim: dimension of source vocabulary
            :param RNN: None, 'LSTM' or 'GRU'
        """
        super(Encoder, self).__init__()
        
        self.source_dim = source_dim
        self.emb_dim = 100
        self.hidden_dim = 100
        self.RNN = RNN

        self.embedding = nn.Embedding(self.source_dim, self.emb_dim)
        
        if self.RNN == None:
            self.embedding2hidden = nn.Linear(self.emb_dim, self.hidden_dim, 1)
        elif self.RNN == 'LSTM':
            self.embedding2hidden = nn.LSTMCell(self.emb_dim, self.hidden_dim)
        elif self.RNN == 'GRU':
            self.embedding2hidden = nn.GRUCell(self.emb_dim, self.hidden_dim)

    def init_hidden(self):
        """
        Initialize hiddden layers.
        """
        if self.RNN == None:
            hidden = None
        elif self.RNN == 'LSTM':
            hidden = (Variable(torch.zeros(1, self.hidden_dim)),Variable(torch.zeros(1, self.hidden_dim)))
        elif self.RNN == 'GRU':
            hidden = Variable(torch.zeros(1, self.hidden_dim))
        return hidden
    
    def forward(self, sentence, hidden_lr, hidden_rl):
        """
        Forward step:
            :param sentence: Tensor containing the indices of the words of the source sentence
            :param hidden_lr: the hidden layer for left to right represention
            :param hidden_rl: the hidden layer for right to left represention
        """
        # define the sizes of the sentences
        length = sentence.size(0)
        
        # find all embeddings
        embeddings = []
        for i in range(length):
            embeddings.append(self.embedding(sentence[i]).view(1, -1))
            
        # create bidirectional representations
        # first left to right
        hiddens_lr = torch.zeros(MAX_SENTENCE_LENGTH, self.hidden_dim)
        for i in range(length):
            if self.RNN == None:
                hidden_lr = self.embedding2hidden(embeddings[i])
            else:
                hidden_lr = self.embedding2hidden(embeddings[i], hidden_lr)
                
            if self.RNN == 'LSTM':
                hiddens_lr[i] = hidden_lr[0][0]
            else:
                hiddens_lr[i] = hidden_lr[0]
            
        # then right to left
        hiddens_rl = torch.zeros(MAX_SENTENCE_LENGTH, self.hidden_dim)
        for i in reversed(range(length)):
            if self.RNN == None:
                hidden_rl = self.embedding2hidden(embeddings[i])
            else:
                hidden_rl = self.embedding2hidden(embeddings[i], hidden_rl)
            if self.RNN == 'LSTM':
                hiddens_rl[i] = hidden_rl[0][0]
            else:
                hiddens_rl[i] = hidden_rl[0]
            
        # concat all left to right and right to left into one vector
        # and put it in a matrtix
        hiddens = torch.zeros(MAX_SENTENCE_LENGTH, 2*self.hidden_dim)
        for i in range(length):
            hiddens[i] = torch.cat([hiddens_rl[i], hiddens_lr[i]])

        return hiddens
```

```python

class Decoder(nn.Module):

    def __init__(self, target_dim, dropout, bilinear):
        """
        Initialize Decoder.
            :param target_dim: dimension of target vocabulary
            :param dropout: regularization value
            :param bilinear: boolean for using attention with bilinear function or dot product
        """
        super(Decoder, self).__init__()
        
        self.target_dim = target_dim
        self.hidden_dim = 100
        self.emb_dim = 100
        self.bilinear_bool = bilinear

        self.embedding = nn.Embedding(self.target_dim, self.emb_dim)
        
        self.embedding2hidden = nn.LSTMCell(self.emb_dim+2*self.hidden_dim, 2*self.hidden_dim)
        
        self.dropout = nn.Dropout(dropout)
    
        if self.bilinear_bool:
            self.bilinear = nn.Linear(2*self.hidden_dim, 2*self.hidden_dim, bias=False)
            
        self.hidden2out = nn.Linear(4*self.hidden_dim, target_dim)
        
        self.softmax = nn.Softmax(dim=0)
        self.relu = nn.ReLU()
        self.log_softmax = nn.LogSoftmax(dim=1)

    def init_cell(self):
        """
        Initialize hiddden layers.
        """
        return Variable(torch.zeros(1, 2*self.hidden_dim))

    def forward(self, encoder_outputs, target):
        """
        Forward step:
            :param encoder_outputs: bidirectional representations of the words in the source sentence
            :param target: Tensor containing the indices of the words of the target sentence (can be None)
        """

        # initialize hidden layer for LSTM
        h = torch.mean(encoder_outputs, 0).view(1,-1)
        c = self.init_cell()
        hidden = (h,c)
        
        # prepare encoder_outputs
        encoder_outputs_transposed = encoder_outputs.unsqueeze(0)
        encoder_outputs = torch.t(encoder_outputs).unsqueeze(0)
        
        # initialize start of sentence (SOS)
        prev_word = torch.tensor(0, dtype=torch.long).view(-1, 1)
        
        # initialize context vector
        context_vector = torch.zeros(1, 2*self.hidden_dim)

        # initialize loop values
        outputs = []
        i = 0
        predict_next_word = True
            
        while predict_next_word:
            
            # compute embedding of previous word
            emb = self.embedding(prev_word)
            
            # regularize the embedding
            emb = self.dropout(emb)
            
            # concat the embedding and previous context_vector
            concat = torch.cat((emb.view(1,-1), context_vector.view(1,-1)),1)

            # go through LSTM layer
            hidden = self.embedding2hidden(concat, hidden)
            output = hidden[0]
                
            # attention with bilinear function vs dot product
            if self.bilinear_bool:
                trans = self.bilinear(output.view(1,1,-1))
                attention_score = torch.bmm(encoder_outputs_transposed, trans.view(1,-1,1)).view(-1,1)
            else:
                attention_score = torch.bmm(encoder_outputs_transposed, output.view(1,-1,1)).view(-1,1)

            # compute softmax over attention
            attention_score = self.softmax(attention_score).unsqueeze(0)
            
            # compute the context vector
            context_vector = torch.bmm(encoder_outputs, attention_score)

            # concat the bidirectional representaion and the context vector
            output = torch.cat((output.view(1,-1), context_vector.view(1,-1)),1)
            
            # go over one more layer to get the output size
            output = self.hidden2out(output)

            # set all negative numbers to zero
            output = self.relu(output)
                        
            # compute the softmax over the output
            output = self.log_softmax(output)
            
            # compute current word by taking argmax of the output
            if target is not None:
                prev_word = target[i]
            else:
                prev_word = torch.argmax(output, 1)
    
            # update loop values
            outputs.append(output)
            i += 1
            if prev_word.data.tolist()[0]==1:
                predict_next_word = False
            
        # outputs: list to tensor
        outputs = torch.cat(outputs, 0)
        
        return outputs
```

```python
class NMTmodel(nn.Module):

    def __init__(self, source_dim, target_dim, dropout, RNN, bilinear):
        """
        Initialize Encoder.
            :param source_dim: dimension of source vocabulary
            :param target_dim: dimension of target vocabulary            
            :param dropout: regularization value
            :param RNN: None, 'LSTM' or 'GRU'
            :param bilinear: boolean for using attention with bilinear function or dot product
        """
        super(NMTmodel, self).__init__()

        self.encoder = Encoder(source_dim, RNN)
        self.decoder = Decoder(target_dim, dropout, bilinear)
    
    def forward(self, sentence, target=None):
        """
        Forward step:
            :param sentence: Tensor containing the indices of the words of the source sentence
            :param target: Tensor containing the indices of the words of the target sentence (optional)
        """
        # clear hidden of encoder for each sentence
        hidden_lr = self.encoder.init_hidden()
        hidden_rl = self.encoder.init_hidden()

        # train encoder
        encoder_outputs = self.encoder(sentence, hidden_lr, hidden_rl)

        # train decoder and find loss
        outputs = self.decoder(encoder_outputs, target)

        return outputs
```

```python
def train(source, target, source_val, target_val, source_dim, target_dim, dropout, epochs, path_model, RNN=None, bilinear=False):
    """
    Create and train a machine translation model.
        :param source: source sentences of train set (list of Tensors containing word indices)
        :param target: target sentences of train set (list of Tensors containing word indices)
        :param source_val: source sentences of validation set (list of Tensors containing word indices)
        :param target_val: target sentences of validation set (list of Tensors containing word indices)
        :param source_dim: dimension of source vocabulary
        :param target_dim: dimension of target vocabulary 
        :param dropout: float between 0 and 1
        :param epochs: number of epochs
        :param path_model: path to save or load the model parameters
        :param RNN: None, 'LSTM' or 'GRU' (optional)
        :param bilinear: boolean for using attention with bilinear function or dot product (optional)
    """
    
    print('Running model')
    
    # set the learning rate
    lr = 0.01
    
    source = source[:100]
    target = target[:100]
    
    assert len(source) == len(target), "Dimension of source data does not agree with dimension of target data."
    assert len(source_val) == len(target_val), "Dimension of source data does not agree with dimension of target data."
    
    # define length of sentences
    all_sentences = len(source)
    all_sentences_val = len(source_val)
    
    # create models
    model = NMTmodel(source_dim, target_dim, dropout, RNN, bilinear)
    
    if os.path.exists(path_model):
        # load in model parameters
        model.load_state_dict(torch.load(path_model))

    # define loss function: cross entropy loss
    criterion = nn.CrossEntropyLoss() 

    # define optimize method: stochastic gradient descent 
    optimizer = torch.optim.SGD(model.parameters(), lr=lr)
    
    # define outputs lists as empty
    total_loss = []
    total_loss_val = []
    
    for epoch in range(epochs):
        
        print('Epoch ', epoch+1)
                
        losss = 0
        losss_val = 0
        
        # sentence and target are matrix
        for j in range(all_sentences):
            
            # zero gradient buffers
            optimizer.zero_grad()

            # train decoder and find loss
            outputs = model(source[j], target[j])
            
            # update the loss
            loss = criterion(outputs, target[j].view(-1))
            
            # update total loss
            losss += loss
            
            # backpropagate the error
            loss.backward()

            # update the weights
            optimizer.step()
            
        # save model after each epoch
        torch.save(model.state_dict(), path_model)
            
        for j in range(all_sentences_val):

            # train decoder and find loss
            outputs = model(source_val[j], target_val[j])
            
            # update the loss
            loss = criterion(outputs, target_val[j].view(-1))
            
            # update total loss
            losss_val += loss
        
        # update total loss
        losss_val /= all_sentences_val
        print("loss of val: ", losss_val)
        total_loss_val.append(losss_val)
        losss /= all_sentences
        print("loss: ", losss)
        total_loss.append(losss)

    return total_loss, total_loss_val
```

## Write model into txt file

```python
def write_output(source, output_path, model_path, source_dim, target_dim, dropout, RNN=None, bilinear=False):
        
    all_sentences = len(source)
    
    # load model
    model = NMTmodel(source_dim, target_dim, dropout, RNN, bilinear)
    model.load_state_dict(torch.load(model_path))
    
    with open(output_path, "w+") as f:
    
        for j in range(all_sentences):

            # find translation
            outputs = model(source[j])

            # find the predicted sentence
            sentence = []
            for i in range(len(outputs)-1):
                max_index = torch.argmax(outputs[i],0)
                sentence.append(english.index2word[int(max_index)])
                
            # write predicted sentence
            f.write(' '.join(sentence))
            f.write('\n')
```
